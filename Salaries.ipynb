### What is the name of highest paid person (including benefits)?

```python
Highest_Paid = df[df['TotalPayBenefits'] == df['TotalPayBenefits'].max()]
Highest_Paid
```

```python
Highest_Paid['TotalPayBenefits']
```

### What is the name of lowest paid person (including benefits)?

```python
Lowest_Paid = df[df['TotalPayBenefits'] == df['TotalPayBenefits'].min()]
Lowest_Paid
```

```python
Lowest_Paid['TotalPayBenefits']
```

### What was the average (mean) BasePay of all employees per year (2011-2014)?

```python
Avg_Per_Year = df.groupby('Year')['BasePay'].mean()
Avg_Per_Year
```

### How many Job Titles were represented by only one person in 2013?

```python
df_2013 = df[df['Year'] == 2013]
df_2013.sort_values('Id').head()
```

```python
Counts = df_2013['JobTitle'].value_counts()
Result = (Counts == 1).sum()
Result
```

### How many people have the word Chief in their job title?

```python
Chief_JobTitle = df[df['JobTitle'].str.contains('Chief', case = False, na = False)]
Counts = Chief_JobTitle['JobTitle'].count()
Counts
```

### Is there a correlation between length of the Job Title string and Salary?

```python
df['TitleLength'] = df['JobTitle'].str.len()
df['TitleLength']
```

```python
Correlation = df['TitleLength'].corr(df['TotalPayBenefits'])
Correlation
```

```python
data = df[['TitleLength','TotalPayBenefits']]
data.corr()
```
